# Discrete Fourier Transforms
An introduction to the numerical methods available in Python.

Formula references refer to the book *Waves and Oscillations* by Walter Fox Smith.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Set the number of data points
N = 50

# Define the time array, some extra options here to set the 
# number of points to N and get the sampling time step, dt
t, dt = np.linspace(0,1,num=N,retstep=True)

# Define the y function as a sin, here w = 100 rad/s
y = np.sin(260*t)
```

```python
# lines with points to emphasize the discrete nature of the data
plt.plot(t,y,".-")
plt.ylabel("y(t)")
plt.xlabel("t")
```

```python
# Using formula 8.6.3, we define a list of the frequencies
w = np.arange(0,N)* 2*np.pi/(N*dt)
```

```python
plt.plot(w,np.abs(np.fft.fft(y)))
```

Clearly the frequencies in the right half are incorrect, and we can safely ignore those anyway as we know the amplitudes are mirrored about the N/2 point (for real input functions).

For the following plot, we can simply chop those values out of the plot:

```python
wmax = np.pi/dt
print(wmax)
plt.plot(w,np.abs(np.fft.fft(y)),".-")
plt.xlim(0,wmax)
plt.xlabel("")
plt.ylabel("Cn")
```

We see a nice single spike at w=100.
