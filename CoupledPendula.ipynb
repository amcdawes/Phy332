Using `scipy` features to run ode integration, we explore the numerical solution to coupled oscillators (pendula in this case).

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from scipy.integrate import odeint
```

```python
t = np.linspace(0,20,400)
```

```python
gl = 15
km = 1
def twopendula(y, t):
    # unpack variables:
    x1, dx1, x2, dx2 = y
    
    # Equations of motion:
    v1 = dx1
    a1 = -gl * x1 - km * (x1 - x2)
    v2 = dx2
    a2 = -gl * x2 - km * (x2 - x1)
    return [v1, a1, v2, a2]
```

```python
result = odeint(twopendula, [1, 0, 0, 0], t)
```

```python
plt.subplot(211)
plt.plot(t,result[:,0])
plt.subplot(212)
plt.plot(t,result[:,2],"orange")
```
