```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
omega_d = np.linspace(0,10,200)
```

```python
def A(omega_d, omega_0=5, gamma=1, F0=1, m=1):
    return (F0/m)/(np.sqrt( (omega_0**2 - omega_d**2)**2 + (gamma*omega_d)**2 ))
```

```python
omega_0 = 5
gamma = 2
F0 = 1
m = 1
plt.plot(omega_d,A(omega_d, omega_0, gamma, F0, m))
plt.vlines(omega_0,0,0.2,"red")
plt.ylabel("A")
plt.xlabel("$\omega_d$")
```

What is $\delta$?

```python
def delta(omega_d,gamma=1,omega_0=5):
    return np.arctan2(gamma*omega_d,(omega_0**2 - omega_d**2))
```

```python
plt.plot(omega_d,delta(omega_d))
plt.vlines(omega_0,0,3,"red")
plt.ylabel("$\delta$")
plt.xlabel("$\omega_d$")
```

$\delta$ is a phase shift between the drive frequency and the oscillation frequency. At low frequency drive, the motion matches the drive (i.e. no phase shift). At resonance, there is a $\pi$/2 phase shift, and at high frequencies, the shift approaches $\pi$.
