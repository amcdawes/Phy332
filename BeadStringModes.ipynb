```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 11
x = np.linspace(0,N,N+1)
x2 = np.linspace(0,N)
```

```python
L = N
n = 4
plt.plot(x,np.sin(n*np.pi/L * x),"o-")
plt.plot(x2,np.sin(n*np.pi/L * x2))
#n2 = 3
#plt.plot(x,np.sin(n2*np.pi/L * x),"o-")
```
